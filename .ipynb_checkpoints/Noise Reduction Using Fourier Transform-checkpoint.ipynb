## Explanation Part

### How it works
Fourier Series decompose any periodic signals into the sum of simple sets of oscillating functions i.g. sin and cosine  

The general idea is that the image : $f(x,y)$ of size $M*N$, can be represented in the frequency domain : $F(u,v)$. 


#### 2D Discrete Fourier Transform

$$
F(u,v) = \frac1{MN} \sum_{x=0}^{M-1} \sum_{y=0}^{N-1} f(x,y) \: e^{-2i\pi (\frac{ux}{M} + \frac{vy}{N}) }
$$  

From equation above, we can treat $f(x,y)$ as each pixel on image. Therefore, we can now represent our image as Fourier Series. 

#### From Euler's Formula

$$
e^{i\theta} = \cos{\theta} + i \sin{\theta}
$$

Thus we can turn our FT equations into sum of $sine$ and $cosine$ function:

$$
e^{-2i\pi (\frac{ux}{M} + \frac{vy}{N})} = \cos({2\pi (\frac{ux}{M} + \frac{vy}{N})}) - i \sin({2\pi (\frac{ux}{M} + \frac{vy}{N})})
$$
  
  
$$
F(u,v) = \frac1{MN} \sum_{x=0}^{M-1} \sum_{y=0}^{N-1} f(x,y) \: (\cos({2\pi (\frac{ux}{M} + \frac{vy}{N})}) - i \sin({2\pi (\frac{ux}{M} + \frac{vy}{N})}) )
$$

Now, after we did some magic on our Fourier Domain, we want to transform it back to original image so that we can visualize the image normally. Hence, we need something called Inverse Fourier Transform (in 2D as well).

#### 2D Inverse Fourier Transform

$$
f(x,y) = \sum_{u=0}^{M-1} \sum_{v=0}^{N-1} F(u,y) \: e^{2i\pi (\frac{um}{M} + \frac{vn}{N}) }
$$  

  
### More Explanation
   After we apply the Fourier Transform, we can now visualize our Frequency Domain of our image. Most of the time, the FQ-Domain will have a bright spot in the very center of the image and darker spot spreading out from the center. However, in the picture where there are the periodic noises, the FQ-Domain will show the bright spots or lines throughtout the spectrum representing the periodic signal occurs in the image (aka high peaks if we plot in 3D-plane). Hence, if we remove the right bright spots/line, we can reduce or eventually remove the noises completely from our image. After we remove those peaks, now we have to convert our picture back so that we can visualize it as usual using Inverse Fourier Transform. Done!

## Steps
1. Convert the image from RGB-scale to Gray-scale colorspace. (If needed)  

$$
Y = 0.299R+0.587G+0.114B
$$
  
2. Apply 2D Fourier Transform on the converted image. (Use $np.fft.fft2(\:)$ for 2D-Fourier Tranform)
3. Shift the FT-image so that the frequency spectrum started in the middle of the plane. (Use $np.fft.fftshift(\:)$)
4. Analyse and remove the peaks that might cause the periodic noise from the frequency domain.
5. Apply Inverse Fourier Transform to the FT-image

# Coding

### Miscellaneous

```python
# Convert RGB into Gray-scale
def makeGrayScale(image):
    return np.dot(image,[0.299,0.587,0.114])

# For visualizing purpose, convert FFT into logarithmic scale
def magnitude_spectrum(shifted_fft):
    sft = np.copy(shifted_fft)
    return 20*np.log(np.abs(sft))
    
# Fourier Transform of image, img
def FFT(img):   
    freq_o = fft2(img)
    #Shift zero-freq component to the center
    fshift_o = fftshift(freq_o)
    return fshift_o

# Find maginutde of each pixels in the images
#  z  = a + bi
# |z| = sqrt(a^2 + b^2)
def magnitude(ft):
    return np.sqrt(np.real(ft)**2 + np.imag(ft)**2)

# Inverse Fourier Transform of the transformed image, fimg
def inverseFFT(fimg):
    shiftback = np.real(ifft2(ifftshift(fimg)))
    return shiftback

# plotting tools
def plot12(img,fimg):
    ms_o = magnitude_spectrum(fimg)
    
    plt.figure()
    fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(6,6))
    
    plt.subplot(221)
    plt.title("Original Image")
    plt.imshow(img,cmap='gray')
    
    plt.subplot(222)
    plt.title("Frequency Domain")
    plt.imshow(ms_o,cmap='gray')

    fig.tight_layout()
    plt.show()
```

### Initialize Image

```python
# image : image to be loaded
# show : True\False for showing image and its FQ-Domain
# return [gray-scaled array, FT array]

def load_image(image,show):
    #original image
    img_o = imread.imread(image)
    
    #for RGB array-like
    if(len(np.shape(img_o)) == 3):
        # ignore alpha channel
        img_o = img_o[:,:,:3]
        gray_o = makeGrayScale(img_o)
    else:
        gray_o = img_o

    fshift_o = FFT(gray_o)

    #magnitude spectrum
    ms_o = magnitude_spectrum(fshift_o)
    
    if(show):
        plot12(gray_o,fshift_o)
    return [gray_o,fshift_o]
```

# Part 2 : Analyse the Peaks Automatically
    NOTE : May require your analysis before let the function works for you

### Misc

```python
# def count_peaks(ft,avg):
#     count = sum([1 for i in ft for j in i if j > 2*avg])
#     return count

def spread(fimg,cy,cx,dw,avg,mag):
    ft = np.copy(fimg)
    for i in range(len(mag)):
            for j in range(len(mag[i])):
                if(mag[i,j] > avg) and not (cy-dw<=i<=cy+dw and cx-dw <= j <= cx+dw):
                    ft[i,j] = 1
    return ft

def center(fimg,cy,cx,dw,avg,maxi):
    ft = np.copy(fimg)
    for j in range(cy-dw,cy+dw+1):
            for i in range(cx - dw,cx+dw+1):
                pix = ft[j,i]
                if(maxi-avg < pix < maxi):
                    ft[j,i] = 1
    return ft

def vertical(fimg,cy,cx,dw):
    ft = np.copy(fimg)
    ft[:cy-dw,cx-dw:cx+dw] = 1
    ft[cy+dw:,cx-dw:cx+dw] = 1
    return ft
    
def horizontal(fimg,cy,cx,dw):
    ft = np.copy(fimg)
    ft[cy-dw:cy+dw,:cx-dw] = 1
    ft[cy-dw:cy+dw,cx+dw:] = 1
    return ft
```

### Main function

```python
# img : images from load_image() : [gray-scaled array, FT-array]
# dw : delta from center
# mode : "c" for removing very high peaks near center,
#        "s" for removing distributed peaks
#        "v" using middle vertical line (except center with radius "w")
#        "h" using middle horizontal line (except center with radius "w")
#        "a" using all modes above
def findAvgFreq(img,dw,mode):
    ft = img[1]
    md = list(mode)
    cx = int(len(ft[0])/2)
    cy = int(len(ft)/2)
    
    # magnitiude of each pixels
    mag = magnitude(ft)
    mini = np.min(mag)
    maxi = np.max(mag)
    
    # finding average of the peaks
    n = (len(ft[0]) * len(ft))
    avg = (sum([j for i in ft for j in i])) / n   
    
    # for cases that peaks are far away from center
    if("s" in md):
        ft = spread(ft,cy,cx,dw,avg,mag)
    # for cases that peaks are near the center
    if("c" in md):
        ft = center(ft,cy,cx,dw,avg,maxi)
    # remove using both methods
    if("v" in md):
        ft = vertical(ft,cy,cx,dw)
    if("h" in md):
        ft = horizontal(ft,cy,cx,dw)

    return ft
```
